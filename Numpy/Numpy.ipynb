NumPy
NumPy (Numerical Python) is a Python library used for creating and working with multi-dimensional arrays.

It is the fundamental package for scientific computing in Python.

Features of NumPy:

Provides high-performance multi-dimensional arrays.
Supports mathematical and statistical operations.
Performs calculations faster than Python lists.
Used in Data Science, Machine Learning, and AI applications.
Works efficiently with large amounts of numerical data.

Diiference Between Array and list:
  Array will support only one data type(int or string or boolean)
  List will support different data type

```python
#copy function 
arr=[1,2,3,4,5,6,7,8,9]
arr=np.array(arr)
arr[3:]=100
# copy of the array
arr1=arr.copy()
arr1 
  
```

```python
# The main use of copy() function is if we give = then two arrays are pointing same varible
# Then if you make changes in one variable the other variable also changes that is why we use copy
x=[1,2,3,4,5]
s=x
print("The original array: ",x) 
s[2]=100
print("Modified array: ",s)
print("Modified original array: ",x)
#this is why we use copy 
d=x.copy()
print("Copied array: ",d)
d[1]=234 #The change does not affect the original array
print("The original array: ",x)
print("the change of copy: ",d)
```
